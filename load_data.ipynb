```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
path = '08.csv'
data = pd.read_csv(path)
data.drop('S/N', axis=1, inplace=True)
data = data.set_index('Sector')
data.head()
```

```python
def simplify_data(data_raw):
    data_raw['Lower education'] = data_raw['Junior middle school'] + data_raw['Senior middle school']
    data_raw['Professional technology'] = data_raw['Technical secondary school'] + data_raw['Junior college']
    data_raw['Highly educated'] = data_raw['Master\'s degree'] + data_raw['Doctor\'s degree'] + data_raw['MBA']
    data_raw['total accepted'] = data_raw['Lower education'] + data_raw['Professional technology'] + data_raw['Bachelor\'s degree'] + data_raw['Highly educated'] + data_raw['Unlimited']
    data = data_raw.reindex(['Total demand (Pers.)', 'total accepted', 'Lower education','Professional technology', 'Bachelor\'s degree', 'Highly educated', 'Unlimited'], axis = 1)
    data_raw['Acceptance rate'] = data_raw['total accepted']/ data_raw['Total demand (Pers.)']
    data_raw['Professional relevance'] = (data_raw['total accepted'] - data_raw['Unlimited']) / data_raw['total accepted']
    data_feature = data_raw.reindex(['Acceptance rate', 'Professional relevance'], axis = 1)
#     display(data)
    return data, data_feature
```

```python
data, data_feature = simplify_data(data)
display(data.head(), data_feature.head())
```
