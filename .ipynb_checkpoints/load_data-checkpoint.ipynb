```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '08.csv'
data = pd.read_csv(path)
data.drop('S/N', axis=1, inplace=True)
data = data.set_index('Sector')
data.head()
```

```python
def simplify_data(data):
    data['Lower education'] = data['Junior middle school'] + data['Senior middle school']
    data['Professional technology'] = data['Technical secondary school'] + data['Junior college']
    data['Highly educated'] = data['Master\'s degree'] + data['Doctor\'s degree'] + data['MBA']
    data = data.reindex(['Total demand (Pers.)', 'Lower education','Professional technology', 'Bachelor\'s degree', 'Highly educated', 'Unlimited'], axis = 1)
    return data
```

```python
simplify_data(data).head()
```

```python
data.set_index('Sector')
pro_data = data['Junior middle school'] + data['Senior middle school']
pro_data = pd.DataFrame(pro_data, columns = ['low B'])

pro_data.set_index(data['Sector']).head()
```
